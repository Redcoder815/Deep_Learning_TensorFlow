```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
x = tf.Variable(tf.range(-8.0, 8.0, 0.1), dtype=tf.float32)
y = tf.nn.relu(x)
plt.figure(figsize=(5,2.5))
plt.plot(x.numpy(), y.numpy())
plt.xlabel('x')
plt.ylabel('relu(x)')
plt.show()
```

```python
with tf.GradientTape() as t:
    y = tf.nn.relu(x)
plt.figure(figsize=(5,2.5))
# compute the derivative of Relu function with respect to x
plt.plot(x.numpy(), t.gradient(y, x).numpy())
plt.xlabel('x')
plt.ylabel('grad of relu')
plt.show()
```

```python
y = tf.nn.sigmoid(x)
plt.figure(figsize=(5,2.5))
plt.plot(x.numpy(), y.numpy())
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.show()
```

```python
with tf.GradientTape() as t:
  y = tf.nn.sigmoid(x)
plt.figure(figsize=(5,2.5))
plt.plot(x.numpy(), t.gradient(y, x).numpy())
plt.xlabel('x')
plt.ylabel('grad of sigmoid')
plt.show()
```

```python
y = tf.nn.tanh(x)
plt.figure(figsize=(5,2.5))
plt.plot(x.numpy(), y.numpy())
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.show()
```

```python
with tf.GradientTape() as t:
  y = tf.nn.tanh(x)
plt.figure(figsize=(5,2.5))
plt.plot(x.numpy(), t.gradient(y, x).numpy())
plt.xlabel('x')
plt.ylabel('grad of tanh')
plt.show()
```
